```python
import pandas as pd

# Load the dataset
df = pd.read_csv('bank.csv', sep=';')

# Display the first 5 rows
df.head()
```

```python
# Shape of the dataset
print("Shape of the dataset:", df.shape)

# Info about columns and data types
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Basic information about the dataset
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the target variable distribution
sns.countplot(x='y', data=df)
plt.title("Distribution of Target Variable ('y')")
plt.xlabel("Subscription Status")
plt.ylabel("Count")
plt.show()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
# Summary statistics for numeric columns
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count plot for target variable
sns.countplot(data=df, x='y')
plt.title('Distribution of Target Variable (y)')
plt.xlabel('Subscribed to Term Deposit')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='job', hue='y')
plt.title('Job vs Subscription')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='marital', hue='y')
plt.title('Marital Status vs Subscription')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='education', hue='y')
plt.title('Education vs Subscription')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='y', y='age')
plt.title('Age Distribution by Subscription')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='y', y='balance')
plt.title('Balance Distribution by Subscription')
plt.ylim(-1000, 10000)  # Limit axis to avoid extreme outliers dominating the view
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='y', y='duration')
plt.title('Call Duration by Subscription')
plt.ylim(0, 1000)  # Optional: to reduce the effect of outliers
plt.show()
```
